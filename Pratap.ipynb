```python
import numpy as np
from PIL import Image
from Class_ImgProcessing import ImageProcessing
```

```python
img = Image.open('ACIK_Logo-removebg.png')
img_arr = np.array(img)
print(img_arr.shape)
ImageProcessing.display_img(img_arr)
```

```python
alpha_channel = img_arr[:, :, 3]
print(alpha_channel)
print(np.sum(alpha_channel == 0))
```

```python
threshold = 30
r, g, b, a = img_arr[:, :, 0], img_arr[:, :, 1], img_arr[:, :, 2], img_arr[:, :, 3]
black_mask = (r <= threshold) & (g <= threshold) & (b <= threshold) & (a >= 250)

new_img_arr = img_arr.copy()
new_img_arr[black_mask] = [220, 220, 220, 255]
ImageProcessing.display_img(new_img_arr)
```

```python
alpha_channel = new_img_arr[:, :, 3]
print(alpha_channel)
print(np.sum(alpha_channel == 0))
```

```python
new_img = Image.fromarray(new_img_arr).convert("RGBA")
new_img.save("ACIK_Website_Logo.png")
```
